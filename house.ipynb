```python
x= data[["number of bedrooms", "number of bathrooms","living area","condition of the house","Number of schools nearby"]]
y = data[["Price"]]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

```

```python

x_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python

from sklearn.tree import DecisionTreeRegressor
param_grid = {
    "criterion" : ["mse", "friedman_mse", "mae"],
    "splitter" : ["best", "random"],
    "max_depth" : [None, 10, 20, 30, 40, 50],
    "min_samples_split" : [2, 5, 10],
    "min_samples_leaf" : [1, 2, 4]
}
```

```python


tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator=tree_model, param_grid=param_grid)
```

```python
grid_tree.fit(x_train,y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds = grid_tree.predict(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predslr = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, predslr)
print("Mean Squared Error:", mse)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

mse = mean_squared_error(y_test, predslr)
rmse = np.sqrt(mse)   # Root Mean Squared Error
mae = mean_absolute_error(y_test, predslr)
r2 = r2_score(y_test, predslr)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
print("RÂ² Score:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel =  RandomForestRegressor()
```

```python
param_gridrfr = {
    "max_depth" : [5,10,15],
    "n_estimators" : [2,3,4,5,6,7,8,9,10]
}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(x_train,y_train)
```

```python
gridrfr.fit(x_train,y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
rfrpredictions = gridrfr.predict(x_test)
```
